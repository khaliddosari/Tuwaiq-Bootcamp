# Part 3 — Validation Functions

Create reusable functions for:

1. cleaning a full name,
2. validating an email address,
3. validating and converting a score,
4. validating a non-empty course code.

## Required Behaviors

### `clean_name(name)`
- remove surrounding spaces,
- replace repeated internal spaces,
- convert to title case.

### `validate_email(email)`
- must contain exactly one `@`,
- must contain a dot after `@`,
- must not contain spaces.

### `validate_score(value)`
- convert the value to `float`,
- accept only values from 0 to 100,
- raise `InvalidScoreError` otherwise.

### `validate_course_code(code)`
- remove spaces,
- convert to uppercase,
- reject an empty value.

````python
def clean_name(name: str) -> str:
    # TODO
    stripped_name = name.strip()
    name_list = stripped_name.split()
    spaced_name = " ".join(name_list)
    name_title = spaced_name.title()
    return name_title


def validate_email(email: str) -> str:
    # TODO
    if email.count("@") != 1:
        raise ValueError("Email must contain exactly one @ symbol.")
    
    domain_part = email.split("@")[1]
    if "." not in domain_part:
        raise ValueError("Email must contain a dot after the @ symbol.")  
    elif " " in email:
        raise ValueError("Email must not contain spaces.")
    else:
        return email


def validate_score(value: Any) -> float:
    # TODO
    try:
        score = float(value)
    except (ValueError, TypeError):
        raise InvalidScoreError("Score must be a number.")
    
    if not MIN_SCORE <= score <= MAX_SCORE:
        raise InvalidScoreError("Score must be between 0 and 100.")
    
    return score


def validate_course_code(code: str) -> str:
    # TODO
    unspaced_code = code.replace(" ", "")
    if unspaced_code == "":
        raise ValueError("A course code must be provided.")
    
    return unspaced_code.upper()
````

# Part 4 — Class `Trainee`

Create a class named `Trainee`.

## Required Attributes

- `trainee_id`
- `name`
- `email`
- `enrolled_courses`
- `attendance`
- `scores`

## Recommended Data Structures

- `enrolled_courses`: set of course codes
- `attendance`: dictionary mapping course code to a set of attended session numbers
- `scores`: dictionary mapping course code to a list of scores

## Required Methods

- `enroll(course_code)`
- `record_attendance(course_code, session_number)`
- `add_score(course_code, score)`
- `average_score(course_code)`
- `to_dict()`
- `from_dict()`
- `__str__()`

````python
class Trainee:
    def __init__(self, trainee_id: int, name: str, email: str):
        # TODO: Validate and store core attributes.
        self.trainee_id = trainee_id
        self.name = clean_name(name)
        self.email = validate_email(email)

        # TODO: Create empty data structures.
        self.enrolled_courses: set[str] = set()
        self.attendance: dict[str, set[int]] = {}
        self.scores: dict[str, list[float]] = {}


    def enroll(self, course_code: str) -> None:
        # TODO
        self.enrolled_courses.add(course_code)


    def record_attendance(self, course_code: str, session_number: int) -> None:
        # TODO
        if course_code not in self.attendance:
            self.attendance[course_code] = set()

        self.attendance[course_code].add(session_number)


    def add_score(self, course_code: str, score: float) -> None:
        # TODO
        validated_score = validate_score(score)

        if course_code not in self.scores:
            self.scores[course_code] = []

        self.scores[course_code].append(validated_score)


    def average_score(self, course_code: str) -> float:
        # TODO
        course_scores = self.scores.get(course_code, [])

        if not course_scores:
            return 0.0

        total = 0.0
        for value in course_scores:
            total += value

        return total / len(course_scores)


    def to_dict(self) -> dict:
        # TODO: Convert sets to JSON-compatible lists.
        attendance_dict = {}
        for course_code, sessions in self.attendance.items():
            attendance_dict[course_code] = list(sessions)

        scores_dict = {}
        for course_code, course_scores in self.scores.items():
            scores_dict[course_code] = list(course_scores)

        return {
            "trainee_id": self.trainee_id,
            "name": self.name,
            "email": self.email,
            "enrolled_courses": list(self.enrolled_courses),
            "attendance": attendance_dict,
            "scores": scores_dict,
        }


    @classmethod
    def from_dict(cls, data: dict) -> "Trainee":
        # TODO: Restore all trainee data.
        trainee = cls(data["trainee_id"], data["name"], data["email"])
        trainee.enrolled_courses = set(data["enrolled_courses"])

        for course_code, sessions in data["attendance"].items():
            trainee.attendance[course_code] = set(sessions)

        for course_code, scores in data["scores"].items():
            trainee.scores[course_code] = list(scores)

        return trainee


    def __str__(self) -> str:
        # TODO
        return f"Trainee({self.trainee_id}, {self.name}, courses = {len(self.enrolled_courses)})"
````

# Part 5 — Class `Course`

Create a class named `Course`.

## Required Attributes

- `code`
- `title`
- `capacity`
- `passing_score`
- `minimum_attendance`
- `schedule`
- `enrolled_trainees`

Use a tuple for `schedule` and a set for `enrolled_trainees`.

## Required Methods

- `has_space()`
- `enroll_trainee(trainee_id)`
- `is_enrolled(trainee_id)`
- `to_dict()`
- `from_dict()`
- `__str__()`

````python
class Course:
    def __init__(
        self,
        code: str,
        title: str,
        capacity: int,
        passing_score: float = DEFAULT_PASSING_SCORE,
        minimum_attendance: float = DEFAULT_MIN_ATTENDANCE,
        schedule: tuple[str, ...] = (),
    ):
        # TODO
        self.code = validate_course_code(code)
        self.title = title.strip()
        self.capacity = capacity
        self.passing_score = passing_score
        self.minimum_attendance = minimum_attendance
        self.schedule = tuple(schedule)
        self.enrolled_trainees: set[int] = set()

    def has_space(self) -> bool:
        # TODO
        return len(self.enrolled_trainees) < self.capacity
    

    def enroll_trainee(self, trainee_id: int) -> None:
        # TODO: Raise CourseFullError when appropriate.
        if not self.has_space():
            raise CourseFullError(f"The course {self.code} is full")
        
        self.enrolled_trainees.add(trainee_id)

    def is_enrolled(self, trainee_id: int) -> bool:
        # TODO
        return trainee_id in self.enrolled_trainees


    def to_dict(self) -> dict:
        # TODO
        return {
            "code": self.code,
            "title": self.title,
            "capacity": self.capacity,
            "passing_score": self.passing_score,
            "minimum_attendance": self.minimum_attendance,
            "schedule": list(self.schedule),
            "enrolled_trainees": list(self.enrolled_trainees),
        }


    @classmethod
    def from_dict(cls, data: dict) -> "Course":
        # TODO
        course = cls(
            data["code"],
            data["title"],
            data["capacity"],
            data["passing_score"],
            data["minimum_attendance"],
            tuple(data["schedule"]),
        )
        course.enrolled_trainees = set(data["enrolled_trainees"])
        return course


    def __str__(self) -> str:
        # TODO
        return f"Course({self.code}, {self.title}, enrolled = {len(self.enrolled_trainees)}/{self.capacity})"
````

# Part 6 — Class `TrainingCenter`

This class controls the complete system.

## Required Attributes

- `name`
- `trainees`: dictionary mapping trainee ID to `Trainee`
- `courses`: dictionary mapping course code to `Course`

## Required Methods

### Trainee and course management
- `add_course(course)`
- `register_trainee(trainee)`
- `get_course(course_code)`
- `get_trainee(trainee_id)`
- `enroll_trainee(trainee_id, course_code)`

### Data recording
- `record_attendance(trainee_id, course_code, session_number)`
- `record_score(trainee_id, course_code, score)`

### Calculations
- `attendance_percentage(trainee_id, course_code)`
- `course_status(trainee_id, course_code)`
- `certificate_eligible(trainee_id, course_code)`

### Reports
- `trainee_report(trainee_id)`
- `course_report(course_code)`
- `overall_summary()`

### Files
- `save_json(path)`
- `load_json(path)`
- `export_course_report_csv(course_code, path)`

````python
class TrainingCenter:
    def __init__(self, name: str):
        self.name = name
        self.trainees: dict[int, Trainee] = {}
        self.courses: dict[str, Course] = {}


    def add_course(self, course: Course) -> None:
        # TODO
        self.courses[course.code] = course


    def register_trainee(self, trainee: Trainee) -> None:
        # TODO: Prevent duplicate trainee IDs.
        if trainee.trainee_id in self.trainees:
            raise DuplicateTraineeError(f"Trainee {trainee.trainee_id} is already registered.")

        self.trainees[trainee.trainee_id] = trainee


    def get_course(self, course_code: str) -> Course:
        # TODO: Raise CourseNotFoundError when needed.
        validated_code = validate_course_code(course_code)
        if validated_code not in self.courses:
            raise CourseNotFoundError(f"Course {validated_code} does not exist.")

        return self.courses[validated_code]


    def get_trainee(self, trainee_id: int) -> Trainee:
        # TODO: Raise TraineeNotFoundError when needed.
        if trainee_id not in self.trainees:
            raise TraineeNotFoundError(f"Trainee {trainee_id} does not exist.")

        return self.trainees[trainee_id]


    def enroll_trainee(self, trainee_id: int, course_code: str) -> None:
        # TODO: Update both Trainee and Course objects.
        trainee = self.get_trainee(trainee_id)
        course = self.get_course(course_code)
        course.enroll_trainee(trainee_id)
        trainee.enroll(course.code)


    def record_attendance(self, trainee_id: int, course_code: str, session_number: int) -> None:
        # TODO
        trainee = self.get_trainee(trainee_id)
        course = self.get_course(course_code)

        trainee.record_attendance(course.code, session_number)


    def record_score(self, trainee_id: int, course_code: str, score: float) -> None:
        # TODO
        trainee = self.get_trainee(trainee_id)
        course = self.get_course(course_code)

        trainee.add_score(course.code, score)


    def attendance_percentage(self, trainee_id: int, course_code: str) -> float:
        # TODO
        trainee = self.get_trainee(trainee_id)
        course = self.get_course(course_code)

        attended_sessions = trainee.attendance.get(course.code, set())
        total_sessions = len(course.schedule)

        return len(attended_sessions) / total_sessions * 100


    def course_status(self, trainee_id: int, course_code: str) -> str:
        # TODO: Return Passed, Failed, or Incomplete.
        trainee = self.get_trainee(trainee_id)
        course = self.get_course(course_code)

        course_scores = trainee.scores.get(course.code, [])
        if not course_scores:
            return "Incomplete"

        average = trainee.average_score(course.code)
        attendance = self.attendance_percentage(trainee_id, course.code)
        has_passed = average >= course.passing_score and attendance >= course.minimum_attendance

        return "Passed" if has_passed else "Failed"


    def certificate_eligible(self, trainee_id: int, course_code: str) -> bool:
        # TODO
        course = self.get_course(course_code)
        if not course.is_enrolled(trainee_id):
            return False

        return self.course_status(trainee_id, course.code) == "Passed"


    def trainee_report(self, trainee_id: int) -> dict:
        # TODO
        trainee = self.get_trainee(trainee_id)

        courses_report = {}
        for course_code in sorted(trainee.enrolled_courses):
            courses_report[course_code] = {
                "average": round(trainee.average_score(course_code), 2),
                "attendance": round(self.attendance_percentage(trainee_id, course_code), 2),
                "status": self.course_status(trainee_id, course_code),
                "certificate_eligible": self.certificate_eligible(trainee_id, course_code),
            }

        return {
            "trainee_id": trainee.trainee_id,
            "name": trainee.name,
            "email": trainee.email,
            "courses": courses_report,
        }


    def course_report(self, course_code: str) -> list[dict]:
        # TODO
        course = self.get_course(course_code)

        report_rows = []
        for trainee_id in course.enrolled_trainees:
            trainee = self.get_trainee(trainee_id)
            report_rows.append({
                "trainee_id": trainee.trainee_id,
                "trainee_name": trainee.name,
                "average_score": round(trainee.average_score(course.code), 2),
                "attendance_percentage": round(self.attendance_percentage(trainee_id, course.code), 2),
                "status": self.course_status(trainee_id, course.code),
                "certificate_eligible": self.certificate_eligible(trainee_id, course.code),
            })

        report_rows.sort(key=lambda row: row["average_score"], reverse=True)

        return report_rows


    def overall_summary(self) -> dict:
        # TODO
        total_enrollments = 0
        passed_records = 0
        failed_records = 0
        incomplete_records = 0
        certificates_eligible = 0

        for trainee_id, trainee in self.trainees.items():
            for course_code in trainee.enrolled_courses:
                total_enrollments += 1

                status = self.course_status(trainee_id, course_code)
                if status == "Passed":
                    passed_records += 1
                elif status == "Failed":
                    failed_records += 1
                else:
                    incomplete_records += 1

                if self.certificate_eligible(trainee_id, course_code):
                    certificates_eligible += 1

        top_course = ""
        top_average = 0.0
        for course_code, course in self.courses.items():
            course_averages = []
            for trainee_id in course.enrolled_trainees:
                trainee = self.get_trainee(trainee_id)
                if trainee.scores.get(course_code, []):
                    course_averages.append(trainee.average_score(course_code))

            if course_averages:
                course_mean = float(np.mean(course_averages))
                if course_mean > top_average:
                    top_average = course_mean
                    top_course = course.title

        return {
            "training_center": self.name,
            "courses": len(self.courses),
            "registered_trainees": len(self.trainees),
            "total_enrollments": total_enrollments,
            "passed_records": passed_records,
            "failed_records": failed_records,
            "incomplete_records": incomplete_records,
            "certificates_eligible": certificates_eligible,
            "top_course": top_course,
        }


    def save_json(self, path: Path) -> None:
        # TODO
        courses_data = []
        for course in self.courses.values():
            courses_data.append(course.to_dict())

        trainees_data = []
        for trainee in self.trainees.values():
            trainees_data.append(trainee.to_dict())

        data = {
            "name": self.name,
            "courses": courses_data,
            "trainees": trainees_data,
        }

        with open(path, "w", encoding="utf-8") as file:
            json.dump(data, file, indent=2)


    @classmethod
    def load_json(cls, path: Path) -> "TrainingCenter":
        # TODO
        with open(path, "r", encoding="utf-8") as file:
            data = json.load(file)

        center = cls(data["name"])
        for course_data in data["courses"]:
            center.add_course(Course.from_dict(course_data))

        for trainee_data in data["trainees"]:
            center.register_trainee(Trainee.from_dict(trainee_data))

        return center


    def export_course_report_csv(self, course_code: str, path: Path) -> None:
        # TODO
        report_rows = self.course_report(course_code)
        field_names = [
            "trainee_id",
            "trainee_name",
            "average_score",
            "attendance_percentage",
            "status",
            "certificate_eligible",
        ]

        with open(path, "w", newline="", encoding="utf-8") as file:
            writer = csv.DictWriter(file, fieldnames=field_names)
            writer.writeheader()
            for row in report_rows:
                writer.writerow(row)
````

# Part 12 — NumPy Course Analysis

   - minimum,
   - maximum,
   - standard deviation,

````python
selected_course_code = "PY101"

# TODO: Collect averages for the selected course.
selected_course = center.get_course(selected_course_code)

selected_averages = []
for trainee_id in sorted(selected_course.enrolled_trainees):
    trainee = center.get_trainee(trainee_id)
    if trainee.scores.get(selected_course_code, []):
        selected_averages.append(trainee.average_score(selected_course_code))

# TODO: Convert averages to a NumPy array.
averages_array = np.array(selected_averages)

# TODO: Calculate and print statistics.
print(f"Averages in {selected_course_code}:", averages_array)
print(f"Mean: {averages_array.mean():.2f}")
print(f"Minimum: {averages_array.min():.2f}")
print(f"Maximum: {averages_array.max():.2f}")
print(f"Standard deviation: {averages_array.std():.2f}")

# TODO: Use Boolean filtering to select passing averages.
passing_averages = averages_array[averages_array >= selected_course.passing_score]
print("Passing averages:", passing_averages)
print("Passing trainees:", len(passing_averages))
````
