</details>

````python
def merge_totals(first: dict, second: dict) -> dict:
    # TODO: Return a new merged dictionary.
    merged = first.copy()
    for key, value in second.items():
        if key in merged:
            merged[key] += value
        else:
            merged[key] = value
    return merged

dict_a = {"A": 12, "B": 8}
dict_b = {"B": 7, "C": 15}

print(merge_totals(dict_a, dict_b))
````

## Exercise 7: Palindrome Sentence

**Level:** Intermediate+  
**Learning objective:** Clean punctuation and spaces before testing symmetry.

### Task
Return `True` if the sentence is a palindrome after ignoring case and non-alphanumeric characters.

### Given data
```python
sentence = "Was it a rat I saw?"
```

### Expected output
```text
True
```

<details>
<summary><strong>Hint</strong></summary>

`str.isalnum()` can identify letters and numbers. Compare the cleaned text with its reverse.

</details>

````python
def is_sentence_palindrome(sentence: str) -> bool:
    # TODO: Build a normalized string and test it.
    normalized = sentence.replace(" ", "").lower()
    for char in normalized:
        if not char.isalnum():
            normalized = normalized.replace(char, "")
    return normalized.isalnum() and normalized == normalized[::-1]


print(is_sentence_palindrome("Was it a rat I saw?"))
````

## Exercise 19: Business Days Calculator

**Level:** Intermediate+  
**Learning objective:** Count weekdays while excluding weekends.

### Task
Count business days from Monday 6 July 2026 through Sunday 12 July 2026, inclusive.

### Given data
```python
start = date(2026, 7, 6)
end = date(2026, 7, 12)
```

### Expected output
```text
5
```

<details>
<summary><strong>Hint</strong></summary>

Advance one day at a time and use `.weekday() < 5`.

</details>

````python
from datetime import date, timedelta

def count_business_days(start: date, end: date) -> int:
    # TODO: Count Monday through Friday, inclusive.
    count = 0
    current_date = start

    while current_date <= end:
        if current_date.weekday() < 5:
            count += 1
        current_date += timedelta(days=1)

    return count

print(count_business_days(date(2026, 7, 6), date(2026, 7, 12)))
````

## Exercise 22: Singly Linked List

**Level:** Advanced  
**Learning objective:** Build and traverse a basic linked data structure.

### Task
Implement append and conversion to a Python list.

### Given data
```python
linked.append(10)
linked.append(20)
linked.append(30)
```

### Expected output
```text
[10, 20, 30]
```

<details>
<summary><strong>Hint</strong></summary>

Each node stores a value and a reference to the next node.

</details>

````python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        # TODO: Add a node at the end.
        new_node = Node(value)
        if not self.head:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def to_list(self):
        # TODO: Traverse from head to end.
        result = []
        current = self.head
        while current:
            result.append(current.value)
            current = current.next
        return result


linked = SinglyLinkedList()
linked.append(10)
linked.append(20)
linked.append(30)
print(linked.to_list())
````

## Exercise 25: Recursive Binary Search

**Level:** Advanced  
**Learning objective:** Use divide-and-conquer recursion on sorted data.

### Task
Return the index of the target or `-1` when absent.

### Given data
```python
values = [3, 8, 12, 19, 24, 31, 45]
target = 24
```

### Expected output
```text
4
```

<details>
<summary><strong>Hint</strong></summary>

Compare the target with the middle value, then recurse into only one half.

</details>

````python
def binary_search(values: list[int], target: int, low=0, high=None) -> int:
    # TODO: Implement recursively.
    if high is None:
        high = len(values) - 1

    if low > high:
        return -1

    mid = (low + high) // 2
    if values[mid] == target:
        return mid
    elif values[mid] < target:
        return binary_search(values, target, low=mid + 1, high=high)
    else:
        return binary_search(values, target, low=low, high=mid - 1)


print(binary_search([3, 8, 12, 19, 24, 31, 45], 24))
````
