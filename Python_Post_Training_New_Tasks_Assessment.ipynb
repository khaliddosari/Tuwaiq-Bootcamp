### Task 32: Nested Service Report

**Level:** Challenge  
**Skills assessed:** nested dictionaries, aggregation

#### Requirement
Calculate the average response time for each service and identify the fastest.

#### Expected output
```text
Appointments: 210.0 ms
Payments: 150.0 ms
Records: 190.0 ms
Fastest service: Payments
```

#### Acceptance criteria
Store calculated averages in a new dictionary.

````python
services = {
    "Appointments": {"times": [200, 220, 210]},
    "Payments": {"times": [140, 155, 155]},
    "Records": {"times": [180, 195, 195]},
}

# TODO: Calculate averages and fastest service.
averages = {}

for service,details in services.items():
    averages[service]=sum(details["times"])/len(details["times"])
fastest_service = min(averages, key = averages.get)

print(f"Appointments: {averages['Appointments']} ms")
print(f"Payments: {averages['Payments']} ms")
print(f"Records: {averages['Records']} ms")
print(f"Fastest service: {fastest_service}")
````
