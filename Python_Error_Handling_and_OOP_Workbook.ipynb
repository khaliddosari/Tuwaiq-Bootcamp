# 1. Understanding Error Types

## Syntax Error

Python cannot understand the code structure.

```python
if score >= 60
    print("Passed")
```

The missing colon prevents Python from parsing the program.

## Runtime Exception

The code is valid, but an error occurs during execution.

```python
result = 10 / 0
```

This raises `ZeroDivisionError`.

## Logical Error

The code runs, but the result is wrong.

```python
area = 2 * (length + width)
```

This formula calculates perimeter, not area.

`try-except` handles runtime exceptions. It does not automatically fix syntax errors or logical mistakes.

## Common Built-in Exceptions

| Exception | Common cause |
|---|---|
| `ValueError` | Invalid value, such as `int("hello")` |
| `TypeError` | Incompatible type or operation |
| `ZeroDivisionError` | Division by zero |
| `IndexError` | Invalid list position |
| `KeyError` | Missing dictionary key |
| `FileNotFoundError` | File does not exist |
| `AttributeError` | Object does not contain the requested attribute |
| `NameError` | Variable name does not exist |

# 2. Basic `try-except`

```python
try:
    risky_operation()
except SomeException:
    recovery_action()
```

Execution flow:

1. Python runs the `try` block.
2. If no exception occurs, the `except` block is skipped.
3. If the matching exception occurs, the remaining statements inside `try` are skipped.
4. Python runs the matching `except` block.
5. The program continues after the structure.

````python
text = "hello"

try:
    number = int(text)
    print("Converted:", number)
except ValueError:
    print("The value is not a valid integer.")

print("Program continues.")
````

# 4. Multiple `except` Blocks

One operation may fail in different ways.

```python
try:
    result = float(first) / float(second)
except ValueError:
    print("Input must be numeric.")
except ZeroDivisionError:
    print("The denominator cannot be zero.")
```

Python checks the handlers from top to bottom and runs the first matching one.

````python
pairs = [("100", "4"), ("ten", "2"), ("8", "0")]

for first, second in pairs:
    try:
        result = float(first) / float(second)
    except ValueError:
        print("Invalid numeric input")
    except ZeroDivisionError:
        print("Cannot divide by zero")
    else:
        print(result)
````

# 8. The `finally` Block

`finally` runs whether the operation succeeds or fails.

```python
try:
    ...
except SomeException:
    ...
finally:
    print("Operation complete")
```

Typical uses:

- releasing resources,
- restoring temporary state,
- recording completion,
- cleanup actions.

For files, a `with` statement is usually safer than manually closing the file.

````python
def divide(value):
    try:
        result = 100 / value
    except ZeroDivisionError:
        print("Cannot divide by zero")
    else:
        print("Result:", result)
    finally:
        print("Operation finished")


divide(4)
print("---")
divide(0)
````

# 9. Raising Exceptions

- return a valid, normalized value, or

````python
def validate_age(age):
    if age < 0:
        raise ValueError("Age cannot be negative.")

    if age > 120:
        raise ValueError("Age is outside the accepted range.")

    return age


for age in [25, -5, 200]:
    try:
        print("Valid age:", validate_age(age))
    except ValueError as error:
        print("Validation error:", error)
````

# 10. Re-Raising and Exception Chaining

## Re-Raising

```python
except ValueError:
    print("Conversion failed.")
    raise
```

This logs context, then allows the caller to handle the error.

## Exception Chaining

```python
except ValueError as error:
    raise ValueError(
        "Employee ID must be an integer."
    ) from error
```

`from error` preserves the original cause while adding a clearer application-level message.

# 11. Custom Exceptions

Create a custom exception when the failure has an important business meaning.

```python
class CourseFullError(Exception):
    pass
```

Custom exceptions make targeted handling easier and improve readability.

````python
class CourseFullError(Exception):
    pass


def enroll(current_students, capacity):
    if current_students >= capacity:
        raise CourseFullError("The course is full.")

    return current_students + 1


try:
    enroll(30, 30)
except CourseFullError as error:
    print(error)
````

````python
# FINAL INTEGRATION PROJECT WORKSPACE

# TODO: Define custom exceptions.
class DuplicateStudentError(Exception):
    pass

class DuplicateCourseError(Exception):
    pass

class StudentNotFoundError(Exception):
    pass

class CourseNotFoundError(Exception):
    pass

class CourseFullError(Exception):
    pass

class InvalidEmailError(Exception):
    pass


# TODO: Define validation functions.
def validate_student_id(student_id):
    if not isinstance(student_id, int) or student_id <= 0:
        raise ValueError("Student ID must be a positive integer.")
    return student_id

def validate_name(name):
    if not name:
        raise ValueError("Name cannot be empty.")
    return name

def validate_email(email):
    if "@" not in email:
        raise InvalidEmailError("Email must contain '@'.")
    local, _, domain = email.partition("@")
    if "." not in domain or domain.startswith(".") or domain.endswith("."):
        raise InvalidEmailError("Email must contain a dot after '@'.")
    return email

def validate_course_code(code):
    if not code:
        raise ValueError("Course code cannot be empty.")
    return code

def validate_capacity(capacity):
    if not isinstance(capacity, int) or capacity <= 0:
        raise ValueError("Capacity must be a positive integer.")
    return capacity


# TODO: Define Student.
class Student:
    def __init__(self, student_id, name, email):
        self.student_id = validate_student_id(student_id)
        self.name = validate_name(name)
        self.email = validate_email(email)
        self.courses = set()

    def enroll(self, course_code):
        self.courses.add(course_code)

    def summary(self):
        return f"{self.name} ({self.student_id}) - {len(self.courses)} course(s)"

    def __str__(self):
        return f"Student({self.student_id}, {self.name})"


# TODO: Define Course.
class Course:
    def __init__(self, code, title, capacity):
        self.code = validate_course_code(code)
        self.title = title
        self.capacity = validate_capacity(capacity)
        self.students = set()

    def is_full(self):
        return len(self.students) >= self.capacity

    def enroll(self, student_id):
        if self.is_full():
            raise CourseFullError(f"Course {self.code} is full.")
        self.students.add(student_id)

    def __len__(self):
        return len(self.students)

    def __str__(self):
        return f"Course({self.code}, {self.title}, {len(self)}/{self.capacity})"


# TODO: Define TrainingCenter.
class TrainingCenter:
    def __init__(self, name):
        self.name = name
        self.students = {}
        self.courses = {}

    def register_student(self, student):
        if student.student_id in self.students:
            raise DuplicateStudentError(f"Student {student.student_id} already registered.")
        self.students[student.student_id] = student

    def add_course(self, course):
        if course.code in self.courses:
            raise DuplicateCourseError(f"Course {course.code} already exists.")
        self.courses[course.code] = course

    def find_student(self, student_id):
        try:
            return self.students[student_id]
        except KeyError:
            raise StudentNotFoundError(f"Student {student_id} not found.") from None

    def find_course(self, course_code):
        try:
            return self.courses[course_code]
        except KeyError:
            raise CourseNotFoundError(f"Course {course_code} not found.") from None

    def enroll_student(self, student_id, course_code):
        student = self.find_student(student_id)
        course = self.find_course(course_code)
        course.enroll(student_id)
        student.enroll(course_code)

    def summary(self):
        return f"Students: {len(self.students)} | Courses: {len(self.courses)}"

    def process_enrollments(self, requests):
        successful = []
        failed = []
        for student_id, course_code in requests:
            try:
                self.enroll_student(student_id, course_code)
            except (StudentNotFoundError, CourseNotFoundError, CourseFullError) as error:
                failed.append({
                    "request": (student_id, course_code),
                    "error_type": type(error).__name__,
                    "message": str(error),
                })
            else:
                successful.append((student_id, course_code))
        return successful, failed


# TODO:
# 1. Create a training center.
center = TrainingCenter("Future Academy")

# 2. Add at least three students.
center.register_student(Student(101, "Sara", "sara@example.com"))
center.register_student(Student(102, "Ali", "ali@example.com"))
center.register_student(Student(103, "Nora", "nora@example.com"))

# 3. Add at least two courses.
center.add_course(Course("PY101", "Python Basics", capacity=2))
center.add_course(Course("DS101", "Data Science", capacity=5))

requests = [
    (101, "PY101"),
    (102, "PY101"),
    (999, "PY101"),
    (103, "MISSING"),
    (103, "PY101"),
]

# 4. Process the enrollment requests.
# 5. Store successful and failed requests.
successful, failed = center.process_enrollments(requests)

# 6. Print detailed results and the final summary.
print("Students:", len(center.students))
print("Courses:", len(center.courses))
print("Successful enrollments:", len(successful))
print("Failed enrollments:", len(failed))
print(successful)
print(failed)
print(center.summary())
````
