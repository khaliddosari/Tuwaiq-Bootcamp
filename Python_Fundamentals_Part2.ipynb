# 1. Conditional Statements

Conditional statements allow a program to make decisions.

## Basic Syntax

```python
if condition:
    # Runs when the condition is True
elif another_condition:
    # Runs when the previous condition is False
else:
    # Runs when all previous conditions are False
```

## Comparison Operators

| Operator | Meaning |
|---|---|
| `==` | Equal to |
| `!=` | Not equal to |
| `>` | Greater than |
| `<` | Less than |
| `>=` | Greater than or equal to |
| `<=` | Less than or equal to |

## Logical Operators

| Operator | Meaning |
|---|---|
| `and` | Both conditions must be true |
| `or` | At least one condition must be true |
| `not` | Reverses the Boolean result |

## Membership Operators

| Operator | Meaning |
|---|---|
| `in` | Value exists in a collection |
| `not in` | Value does not exist in a collection |

## Worked Example 1 — Simple Decision

````python
temperature = 32

if temperature > 30:
    print("The weather is hot.")
else:
    print("The weather is moderate.")
````

## Worked Example 2 — Multiple Conditions

````python
score = 86

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print("Grade:", grade)
````

## Worked Example 3 — Logical Operators

````python
score = 82
attendance = 88

passed = score >= 60 and attendance >= 75

print("Passed:", passed)
````

## Worked Example 4 — Membership Check

````python
allowed_roles = {"admin", "teacher", "editor"}
user_role = "teacher"

if user_role in allowed_roles:
    print("Access granted")
else:
    print("Access denied")
````

## Conditional Expression

A short condition can be written in one line:

```python
status = "Passed" if score >= 60 else "Failed"
```

Use this form only when it remains readable.

## Common Mistakes

### Using `=` instead of `==`

Incorrect:

```python
if score = 90:
```

Correct:

```python
if score == 90:
```

### Checking conditions in the wrong order

Always check the highest or most specific range first.

### Incorrect indentation

All statements inside a branch must be indented consistently.

# 2. Loops

Loops repeat a block of code.

## `for` Loop

Use a `for` loop to iterate over a sequence:

```python
for item in items:
    print(item)
```

## `range()`

| Form | Example |
|---|---|
| Stop only | `range(5)` |
| Start and stop | `range(1, 6)` |
| Start, stop, step | `range(2, 11, 2)` |

## `while` Loop

Use a `while` loop when repetition depends on a changing condition:

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

## Useful Loop Tools

| Tool | Purpose |
|---|---|
| `break` | Stop the loop immediately |
| `continue` | Skip the rest of the current iteration |
| `enumerate()` | Get both position and value |
| `zip()` | Iterate over related collections |
| Nested loop | Loop inside another loop |

## Worked Example 1 — for Loop with range

````python
for number in range(1, 6):
    print(number, "squared =", number ** 2)
````

## Worked Example 2 — enumerate

````python
courses = ["Python", "SQL", "Git"]

for position, course in enumerate(courses, start=1):
    print(f"{position}. {course}")
````

## Worked Example 3 — while Loop

````python
count = 5

while count >= 1:
    print(count)
    count -= 1

print("Go!")
````

## Worked Example 4 — break and continue

````python
numbers = [5, -2, 8, 0, 12]

for number in numbers:
    if number < 0:
        continue

    if number == 0:
        break

    print("Processed:", number)
````

## Common Mistakes

### Infinite Loop

```python
count = 1

while count <= 5:
    print(count)
```

The value of `count` never changes.

### range Endpoint

`range(1, 5)` produces `1, 2, 3, 4`, not `5`.

### Modifying a List During Iteration

Changing a list while looping through it can cause unexpected results.

# 3. Functions

A function is a reusable block of code.

## Basic Syntax

```python
def function_name(parameters):
    return result
```

## Important Terms

| Term | Meaning |
|---|---|
| Function definition | Code beginning with `def` |
| Parameter | Variable in the definition |
| Argument | Value passed during a call |
| Return value | Result sent back to the caller |
| Local variable | Exists inside the function |
| Default parameter | Used when no argument is provided |

## Why Use Functions?

- Reduce repeated code
- Improve readability
- Make testing easier
- Divide large problems into smaller steps

## Worked Example 1 — Parameters and Return

````python
def calculate_rectangle_area(length: float, width: float) -> float:
    return length * width


result = calculate_rectangle_area(8, 5)
print("Area:", result)
````

## Worked Example 2 — Default Parameter

````python
def greet_student(name: str, course: str = "Python") -> str:
    return f"Welcome {name} to {course}."


print(greet_student("Maha"))
print(greet_student("Sara", "Data Science"))
````

## Worked Example 3 — Multiple Return Values

````python
def score_summary(scores: list[float]) -> tuple[float, float, float]:
    minimum = min(scores)
    maximum = max(scores)
    average = sum(scores) / len(scores)

    return minimum, maximum, average


minimum, maximum, average = score_summary([80, 90, 100])

print("Minimum:", minimum)
print("Maximum:", maximum)
print("Average:", average)
````

## Worked Example 4 — *args and **kwargs

````python
def summarize_numbers(*numbers, **metadata):
    return {
        "count": len(numbers),
        "total": sum(numbers),
        "metadata": metadata,
    }


print(summarize_numbers(10, 20, 30, course="Python", week=2))
````

## Scope

A local variable exists inside a function.

```python
def example():
    local_value = 10
```

Functions should usually receive required data through parameters.

## Common Mistakes

- Printing instead of returning
- Forgetting to call the function
- Using unclear parameter names
- Using mutable default values such as `items=[]`

### Exercise 26: Helper Functions

**Level:** Intermediate+

````python
def calculate_average(scores: list[float]) -> float:
    # TODO
    total = sum(scores)
    average = total / len(scores)
    return average


def determine_status(average: float) -> str:
    # TODO
    if average >= 85:
        return "Passed"
    else:
        return "Failed"


def build_report(name: str, scores: list[float]) -> str:
    # TODO
    avg = calculate_average(scores)
    status = determine_status(avg)
    return f"{name} | Average: {avg} | {status}"


print(build_report("Maha", [80, 85, 90]))
````

## Worked Example 1 — Safe Conversion

````python
raw_values = ["25", "hello", "40"]

for raw_value in raw_values:
    try:
        number = int(raw_value)
    except ValueError:
        print(f"{raw_value!r} is not a valid integer.")
    else:
        print("Converted:", number)
````

## Worked Example 2 — Multiple except Blocks

````python
def safe_divide(first: str, second: str) -> None:
    try:
        result = float(first) / float(second)
    except ValueError:
        print("Invalid numeric input")
    except ZeroDivisionError:
        print("Cannot divide by zero")
    else:
        print("Result:", result)
    finally:
        print("Operation complete")


safe_divide("100", "4")
safe_divide("ten", "2")
safe_divide("8", "0")
````

# 6. Object-Oriented Programming Basics

Object-oriented programming organizes code using classes and objects.

## Class and Object

A class is a blueprint:

```python
class Student:
    pass
```

An object is an instance:

```python
student = Student()
```

## Constructor

```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score
```

## Instance Method

```python
def report(self):
    return f"{self.name}: {self.score}"
```

## Main Concepts

| Concept | Meaning |
|---|---|
| Class | Blueprint |
| Object | Instance |
| Attribute | Data stored in an object |
| Method | Function inside a class |
| Encapsulation | Controlling access to state |
| Inheritance | Building a class from another class |
| Polymorphism | Same interface, different behavior |
| Composition | One object contains another object |

## Worked Example 1 — Basic Class

````python
class Student:
    def __init__(self, name: str, score: float):
        self.name = name
        self.score = score

    def status(self) -> str:
        return "Passed" if self.score >= 60 else "Failed"

    def report(self) -> str:
        return f"{self.name}: {self.score} - {self.status()}"


student = Student("Maha", 92)
print(student.report())
````

## Worked Example 2 — Property Validation

````python
class Temperature:
    def __init__(self, celsius: float):
        self.celsius = celsius

    @property
    def celsius(self) -> float:
        return self._celsius

    @celsius.setter
    def celsius(self, value: float) -> None:
        if value < -273.15:
            raise ValueError("Temperature cannot be below absolute zero.")

        self._celsius = value


temperature = Temperature(25)
print(temperature.celsius)
````

## Worked Example 3 — Inheritance and Overriding

````python
class Notification:
    def send(self, message: str) -> str:
        raise NotImplementedError


class EmailNotification(Notification):
    def send(self, message: str) -> str:
        return f"Email: {message}"


class SMSNotification(Notification):
    def send(self, message: str) -> str:
        return f"SMS: {message}"


for channel in [EmailNotification(), SMSNotification()]:
    print(channel.send("Lecture starts at 10"))
````

## Worked Example 4 — Composition

````python
class Engine:
    def __init__(self, engine_type: str):
        self.engine_type = engine_type

    def start(self) -> str:
        return f"{self.engine_type} engine started"


class Car:
    def __init__(self, engine_type: str):
        self.engine = Engine(engine_type)

    def start(self) -> None:
        print(self.engine.start())
        print("Car is ready")


Car("Electric").start()
````
